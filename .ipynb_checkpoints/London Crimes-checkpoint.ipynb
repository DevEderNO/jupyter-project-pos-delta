```python
import pandas as pd #biblioteca para criação de Dataframes
import numpy as np
import matplotlib.pyplot as plt #plotagem
%matplotlib inline
plt.rcParams['figure.figsize'] = (11,11)
ds = pd.read_csv('london.csv',sep=',')
#Filtro para desconsiderar informações com quantidade zerada.
filterQtdeMZ = ds.qtde > 0
#Aplicando o filtro ao dataset.
ds = ds[filterQtdeMZ]
```

# Apresentação do Data Set

Esses dados contam o número de crimes em dois níveis geográficos diferentes de Londres (LSOA e distrito) por ano, de acordo com o tipo de crime. Inclui dados de 2011 até 2016.


Logo abaixo temos uma pequena apresentação de como os dados estão dispostos no data set.

```python
ds.head()
```

Analisando a correlação dos dados para saber quais dados tem mais afinidade uns com os outros podemos observar que não a uma relação muito forte entre os dados por questão da quantidade de informações dispostas no data set.

```python
ds.corr()
```

A relação mais forte que podemos observar nessa correlação e do ano com a quantidade de crimes cometidos. Levendo em consideração essa informação vamos analisar as seguintes questões:

1) Qual é a mudança no número de incidentes de crime de 2011 para 2016

2) Dos anos de 2011 a 2016 quais foram os anos que mais foram cometidos crimes em Londres.

3) Quais foram os três principais crimes por distrito em 2016?

4) Dos crimes cometidos qual a categoria que é mais recorente nos anos que mais ouveram crimes em Londres.

5) E levando em consideração a resposta da questão 4, em qual mês esse tipo de crime mais ocorrereu.

# Qual é a mudança no número de incidentes de crime de 2011 para 2016

Nesse caso montaremos um histograma para verificar a mudança nos incidentes de crimes durante os anos.

```python
ds.ano.hist(bins=11, color='blue', zorder=2, figsize=(11,7) ) 
plt.title('Histograma - Análise de crimes por ano')
plt.ylabel('Quantidade')
plt.show()
```

Ouve uma mudança de 5 para 11 crimes cometidos um aumento de mais de 100%, sendo que em 2012 ouve um aumento acima da media dos demais anos com 15 crimes cometidos um aumento de 200% na taxa de crime desse ano comparado com o ano anterior.

# Dos anos de 2011 a 2016 quais foram os anos que mais foram cometidos crimes em Londres.

Ainda analisando o grafico anterior pode-se observar que os anos que mais ouveram incidentes de crimes em Londres foram nos anos de 2012 e 2016. Sendo em 2012 o ano que mais ocorreram incidentes.

# Quais foram os três principais crimes por distrito em 2016?

Para respender a essa pergunta vamos montar um grafico de piza para melhor visualizar os crimes cometidos por distrito.

```python
#filtro de ano == 2016.
filterAno = ds.ano == 2016
#aplicando o filtro ao dataset agrupando as informações por destrito e gerando o grafico de pizza.
ds[filterAno].groupby('codigo').categoria_principal.value_counts().plot(kind='pie', title='Categoria principal de crimes em 2012', legend=True, table=True, autopct='%1.1f%%').legend(bbox_to_anchor=(1.9, 1))
```

Os principais três principais crimes por distrito são:

1) **Roubo e Manuseio com 18,2%** em três distritos diferentes somando **54,6%** dos crimes cometidos.

2) **Violência contra a pessoa com 9,1%** em três distritos diferêntes somando **27,3%** dos crimes cometidos.

3) **Roubo e Outras Ofênças Notificáveis com 9,1%** cada uma em cada distrito somando **18,2%** dos crimes cometidos.

# Dos crimes cometidos qual a categoria que é mais recorente nos anos que mais ouveram crimes em Londres.

Como analisado anteriormente os anos que mais ocorreram crimes em Londres foram nos anos de 2012 e 2016 com base nessa informação vamos verificar quais os crimes que mais ocorreram nesses anos.

Primeiro filtramos os dados pelos anos de 2012 e 2016 e depois agupamos por cateria_principal e somamos as quantidades para obter a tabela abaixo:

```python
#filtro pelo anos que mais ouverram ocorrência de crimes.
filterAnoMIC = (ds.ano == 2012) | (ds.ano == 2016)
#aplicando o filtro agrupando por categoria e somando a quantidade de cada ocorrência.
ds[filterAnoMIC].groupby('categoria_principal').qtde.sum()
```

Com esta tabela podemos observar que o crime mais cometidos nesses anos foi o crime de **Roubo e Manuseio com 29 ocorrências** nesses dois anos. 

# E levando em consideração a resposta da questão 4, em qual mês esse tipo de crime mais ocorrereu.

Sabendo que o incidente que mais ocorreu em 2012 e 2016 vamos descobrir qual foi o mês em que eles mais ocorrêram.
Para isso temos que realizar novamente os filtros de ano e do incidente que mais ocorreu levantado na questão anterior.

```python
#filtro pela categoria de incidente que mais ocorreu.
filterCategoria = ds.categoria_principal == 'Roubo e Manuseio'
#criando um novo dataset filtrado para poder manusear os dados de uma maneira melhor.
dsFiltrado = ds[filterAnoMIC & filterCategoria]
#removendo a coluna de ano para poder agrupar o dados pelo mês e somando a quantidade da ocorrência Roubo e -
#Manuseio em cada mês
dsFiltrado.drop('ano',axis=1).groupby('mês').sum()
```

Com a tabela acima podemos observar que o mês com maior ocorrencia de crimes foi no mês de **Feverreiro com 6 ocorrências**
